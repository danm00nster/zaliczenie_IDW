```python
from numpy import inner
#wczytaj i zbiory i usuń wiersze od 40 do 70
parking=pd.read_csv('ML-datasets/parking.csv')
rating=pd.read_csv('ML-datasets/rating.csv')
display(parking)
display(rating)
parking.drop(parking.loc[40:70].index, inplace=True)
rating.drop(rating.loc[40:70].index, inplace=True)

#usuń ostatnią kolumnę ze zbioru rating i zapisz jako rating 2
rating2=rating.iloc[:, :-1]
display(rating)
display(rating2)

#połącz ze sobą te zbiory używając inner join w funkcji merge i join
merged_parking_rating = pd.merge(parking, rating, left_on='placeID', right_on='placeID', how = 'inner')
display(merged_parking_rating)
display(rating.keys())

joined_parking_rating=parking.set_index('placeID').join(rating.set_index('placeID'), how = 'inner')
print(joined_parking_rating)

#użyj concat łącząc po kolumnach i wyświetl ile jest pustych wartości - wyświetlam dla wszystkich
concated_parking_rating= pd.concat([parking, rating])
for col in concated_parking_rating.columns:
  print(f'ilość wartośći NAN w kolumnie {col} {concated_parking_rating[col].isna().sum()}')

#posortuj rosnąco po kolumnie food_rating

merged_parking_rating.sort_values(by='food_rating', inplace=True)
display(merged_parking_rating)
#użyj sumy skumulowanej po kolumnie service_rating
rating['service_rating'].cumsum()
```

**Zadanie 3**

wykonane w zadaniu transformacje poza wykorzystaniem w ML pozwalają np. zwizualizować dane o różnych skalach na jednym wykresie. LabelEncoder i OneHotEcoder pozwalają zamienić dane tekstowe na dane numeryczne.

```python

from sklearn import preprocessing
#normalizacja i standaryzacja na zbiorze iris dl 4 pierwszych kolumn
from sklearn.preprocessing import StandardScaler, Normalizer, LabelEncoder, OneHotEncoder
print(iris)
norm=Normalizer()
iris_norm=norm.fit_transform(iris.iloc[:,0:4])
iris_norm=pd.DataFrame(iris_norm)

#przywrócenie oryginalnych nagłówków i dodanie sufixu _norm
iris_norm.columns=iris.columns[0:4] + '_norm'
print(iris_norm)
sc=StandardScaler()
iris_norm_std=sc.fit_transform(iris_norm)
iris_norm_std=pd.DataFrame(iris_norm_std)

#przywrócenie oryginalnych nagłówków i dodanie sufixu _std
iris_norm_std.columns=iris.columns[0:4] + '_std'
print(iris_norm_std)
#dołączenie kolumny species to dataframe po normalizacji i standaryzacji 
iris2=pd.concat([iris_norm_std,iris['species']], axis=1)
print(iris2.iloc[:,3:5])
print("->>>>>> iris2")
print(iris2)

#label encoder, zamiana nazw na cyfry
le=preprocessing.LabelEncoder()
iris2_le=iris2
iris2_le['species']=le.fit_transform(iris2_le['species'])
print(iris2_le)

#one hot encoder
ohe=OneHotEncoder()
iris2_ohe=iris2.join(pd.DataFrame(ohe.fit_transform(iris2[['species']]).toarray()))
print(iris2_ohe)
iris2_ohe.drop(columns=['species'],inplace=True)
print(iris2_ohe)
```

**Zadanie 4**

Wczytaj dowolny zbiór danych i za pomocą matplotlib dokonaj wykresu punktowego, liniowego oraz histogramu


```python
import numpy as np
import matplotlib.pyplot as plt
weather=pd.read_csv('ML-datasets/weather.csv', parse_dates= True, index_col='date')
print(weather )
```

```python
fig, ax=plt.subplots(figsize=(15,10))

ax.plot(weather.index, weather['temp_max'])
ax.set_xlabel('Date')
ax.set_ylabel('Temp')
ax.grid()
plt.show()
```

```python
le=preprocessing.LabelEncoder()
weather['weather_enc']=le.fit_transform(weather['weather'])
print(weather['weather'].groupby(weather['weather_enc']).unique())
fig, ax = plt.subplots(figsize=(15,10))
scatter=ax.scatter(weather['temp_max'], weather['temp_min'], c= weather['weather_enc'], s=200, alpha=0.4 )
ax.set_xlabel('Temp MAX')
ax.set_ylabel('Temp MIN')
ax.legend(*scatter.legend_elements(),loc='lower right', title="Weather Type", title_fontsize=14, fontsize=12,markerscale=3)
ax.grid()
plt.show()
```

```python



import datetime as dt
fig, ax=plt.subplots(figsize=(15,10))
weather['year']=pd.to_datetime(weather.index)
weather['year']=weather['year'].dt.year
ax.plot(weather['year'], weather['temp_max'])
ax.set_xlabel('Date')
ax.set_ylabel('Temp')
ax.grid()
plt.show()

```
