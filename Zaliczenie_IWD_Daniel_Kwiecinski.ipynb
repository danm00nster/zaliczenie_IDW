```python
import pandas as pd
```

```python
iris=pd.read_csv('ML-datasets/iris.csv')
print(iris)
print(iris.columns)
print('-->wszystkie wiersze, dwie pierwsze kolumny')
display(iris.iloc[0:,[0,1]])

print('-->statystyki opisowe')
print(iris.describe())
kolumny=iris.columns[0:-1]
for col in kolumny:
  print(f'statystyki opisowe dla > {col} <')
  print('_______________________________________')
  print(f'  średnia:  {iris[col].mean().round(1)}')
  print(f'  minimum:  {iris[col].min()}')
  print(f'  maksimum: {iris[col].max()}')
  print(f'  std:      {iris[col].std().round(1)}')

print('-->wartości 5 kolumny i wierszy od 3 do 5')
display(iris.iloc[2:5, [4]])

print('-->ostatni wiersz i dwie ostatnie kolumny')
display(iris.iloc[-1,[-2,-1]])

print('-->wartości dla kolumny sepal length > 3')
#wszystkie kolumny dla sepal length >3
display(iris[iris['sepal length'] >3])
#tylko kolumna sepal length z warunkiem >3
display(iris[iris['sepal length'] >3]['sepal length'])
```
